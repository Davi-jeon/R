##### [3]심볼 컬러 설정하기

```r
plot(0:6,0:6, pch=22, type="n", xaxt="n", yaxt="n", ylab="심볼 테두리 번호", xlab="심볼의 배경색 번호")
axis(1, at=1:5) # x축의 범위를 1~5까지 해라. 1=x축
axis(2, at=1:5) # y축의 범위를 1~5까지 해라. 1=x축

points(1:5,rep(1,5),pch=c(21,22,23,24,25),bg=1:5,col=1)         #심볼의 배경색을 1부터 5까지로 설정하고, 심볼 테두리색을 1로 한다.
points(1:5,rep(2,5),pch=c(21,22,23,24,25),bg=1:5,col=1, lwd=2)  #심볼의 배경색을 1부터 5까지로 설정하고, 심볼 테두리색을 1, 심볼의 넓이를 2로 한다. 
points(1:5,rep(3,5),pch=c(21,22,23,24,25),bg=1:5,col=2, lwd=2) #심볼의 배경색을 1부터 5까지로 설정하고, 심볼 테두리색을 2, 심볼의 넓이를  2로 한다.
points(1:5,rep(4,5),pch=c(21,22,23,24,25),bg=1:5,col=3, lwd=2) #심볼의 배경색을 1부터 5까지로 설정하고, 심볼 테두리색을 3, 심볼의 넓이를 2로 한다.
```

##### [9]잘 추정된 추정식과 잘못 추정된 점을 비교

```r
plot(cars, xlim = c(0,25))
abline(a=-5, b=3.5, col="red")

abline(h=mean(cars$dist), lty=2, col="blue")
abline(v=mean(cars$speed), lty=2, col="green")

#추정선(빨간선)이 평균을 지나지 않으므로, 변경이 필요하다.
```

##### [10]데이터를 잘 설명하는 추정선 그리기 
- lm함수 이용 - 평균을 지나는 회귀선을 그린다.

```r
plot(cars, xlim =c(0,25))

# 평균을 지나지 않는 선
abline(a=-5, b=3.5, col = "green")

# 평균을 지나는 최소자승법에 의해서 추정된 회귀식
abline(lm(cars$dist ~ cars$speed), col='red')

# 독립변수 x의 평균, 종속변수 y의 평균을 도식
abline(h=mean(cars$dist), lty=2, col="blue")
abline(v=mean(cars$speed), lty=2, col="green")
```

#### [11] x축에 도식된 그림에 범례그리기
- x축에 변수명이 오고, 각 변수의 값이 y축에 도식된 그림을 그린다. 
- MASS 패키지에서 parcoord()함수이용

```r
library(MASS)
```

```r
parcoord(iris[1:4], col=as.integer(factor(iris$Species))) # col로 해당 값의 species를 정해줌!

#범례
legend("topleft",legend=levels(iris$Species),
      col=1:length(levels(factor(iris$Species))),
       lty=1:length(levels(factor(iris$Species)))
       )
```
