##### voice.part별 히스토그램
- histogram
- QQplot
- side-by-side 차트

```r
# histrogram
# Soprano2는 키가 주로 60에서 70사이에 분포해 다른 voice.part보다 더 키가 작은 singer이 많다.

histogram(~ height | voice.part,  # #or를 사용해 패널을 추가한다.  height = 연속형 변수, voice.part = 레벨이 있는 변수
         data= singer,
         aspect = 1,
         layout= c(2,4),
         nint= 15,
         xlab = "Height (inches)",
         ylab = "Youngseung"
         )
```

```r
# Quantile=Quantile plot
# 정규성을 따르는지 Q-Q plot로 보았을 때, 대체적으로 점들이 선에 분포되어 있고, Soprano1 와 Alto2가 선을 약간 벗어나 있다. 

qqmath(~ height | voice.part,  #or를 사용해 패널을 추가한다.  height = 연속형 변수, voice.part = 레벨이 있는 변수
      data= singer,
      aspect = 1,
      layout=c(2,4),
      prepanel = prepanel.qqmathline,
      panel = function(x, ...){
          panel.grid()
          panel.qqmathline(x, ...)
          panel.qqmath(x, ...)
      },
      xlab = "Unit Normal Quantile",
      ylab = "Height (inches)")
```

```r
#side-by-side차트

barchart(Freq ~ height, groups = voice.part,
        data = singer,
        stack = TRUE, horizontal = FALSE,
        par.settings = list(superpose.plygon = list(col=rainbow(8))),
        auto.key = list(x=.6, y=.7, cornoer=c(0,0)))
```
