###  1. 관련 함수 정리

- sort - 데이터를 순서대로 정렬
- order  - 요인의 순위를 매기는 것 (order와 ordered는 다르다.)
- ordered - 레벨의 순서를 정하는 함수


예제

```r
# ordered
vScore <- c(30, 10, 40, 50, 20, 20, 10, 20, 50)
fScore3 <- ordered(vScore) #레벨이 생기고 레벨의 순위가 생긴다
fScore3
```

```r
# order
vScore <- c(30, 10, 40, 50, 20, 20, 10, 20, 50)
order(vScore)  #각 값의 순위를 알려줌(작은 순서) - 데이터의 위치값을 출력한다. 
```

```r
# order - 서열을 산출한 순위가 아니라 값으로 보고자 하면
vScore[order(vScore)]
```

```r
# sort
vScore <- c(30, 10, 40, 50, 20, 20, 10, 20, 50)
sort(vScore)  #데이터를 순서대로 정렬한다.
```

```r
# order()의 dcreasing = T   -   내림차순으로 정렬된다.
vScore_d <- c(30, 10, 40, 50, 20, 20, 10, 20, 50)

order(vScore_d)
order(vScore_d, decreasing = T)
```

##### 3. data를 특정 컬럼으로 정렬, na값 삭제
- month, day로 정렬하고, na값 삭제

```r
aqm <- melt(airquality, id=c("month","day"), na.rm = TRUE)  
# "month","day"는 양적자료이나 ID로 지정한다.
# month와 day를 id로 삼고, variable이란 행에 month, day외엔 값으로 들어가고, value열에 그에 해당하는 값이 들어간다. 
```

##### 4. 그룹화 

```r
#acst(데이터셋, fomula - x값 ~ y값, 값의 형태, 값 - subset = .(variable == " "))

acast(aqm, variable ~ month, mean, subset = .(variable == "ozone"))

acast(aqm, variable ~ month, mean, subset = .(variable == "solar.r"))

acast(aqm, variable ~ month, mean, subset = .(variable == "wind"))

acast(aqm, variable ~ month, mean, subset = .(variable == "temp"))

#결과 - ozone의 값은 temp와 밀접한 관계가 있다고 판단할 수 있다. 
```
